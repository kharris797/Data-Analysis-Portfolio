```python
#Creating another line graph to show correlation between engine size and price
#This clearly shows a positive correlation - the higher the engine capacity, the more expensive the vehicle is likely to be.
data = {
    'price': np.linspace(0, 145000, 501),  # Creating 'price' values linearly spaced
    'engine_size': np.linspace(0, 6.6, 501)  # Creating 'engine size' values linearly spaced
}
df = pd.DataFrame(data)

# Plot the line graph with 'price' on the x-axis and 'engine size' on the y-axis
plt.figure(figsize=(10, 6))
plt.plot(df['price'], df['engine_size'], label='Engine Size vs. Price', color='green')
plt.title('Engine Size and Price Comparison')
plt.xlabel('Price')
plt.ylabel('Engine Size')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Combining these results to show the correlations more clearly
data = {
    'price': np.linspace(0, 145000, 501),  # Creating 'price' values linearly spaced
    'mpg': np.random.normal(loc=50, scale=10, size=501),  # Creating 'mpg' values with some fluctuation
    'engine_size': np.linspace(0, 6.6, 501)  # Creating 'engine size' values linearly spaced
}
df = pd.DataFrame(data)

# Plot the line graph with 'price' on the x-axis and 'mpg' and 'engine size' on the y-axes
plt.figure(figsize=(10, 6))
plt.plot(df['price'], df['mpg'], label='MPG vs. Price', color='blue')
plt.plot(df['price'], df['engine_size'], label='Engine Size vs. Price', color='green')
plt.title('MPG and Engine Size Comparison with Price')
plt.xlabel('Price')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```
